```python
clean_df["std_og_t2o_mou"].fillna(clean_df["std_og_t2o_mou"].median(),inplace=True)
clean_df["loc_ic_t2o_mou"].fillna(clean_df["loc_ic_t2o_mou"].median(),inplace=True)
clean_df["loc_og_t2o_mou"].fillna(clean_df["loc_og_t2o_mou"].column(),inplace=True)
# date 
clean_df["last_date_of_month_8"].fillna(clean_df["last_date_of_month_8"].mode()[0],inplace=True)
clean_df["date_of_last_rech_6"].fillna(clean_df["date_of_last_rech_6"].mode()[0],inplace=True)
clean_df["last_date_of_month_9"].fillna(clean_df["last_date_of_month_9"].mode()[0],inplace=True)
clean_df["date_of_last_rech_7"].fillna(clean_df["date_of_last_rech_7"].mode()[0],inplace=True)
clean_df["date_of_last_rech_8"].fillna(clean_df["date_of_last_rech_8"].mode()[0],inplace=True)
```
